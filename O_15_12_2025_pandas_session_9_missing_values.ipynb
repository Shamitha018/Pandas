```python
import pandas as pd
import numpy as np
diction1= {'Names': ['Shamitha', 'Lasya', 'Ridhi', np.nan],
           'Age': [np.nan,22,18, 30],
           'City': ['Hyd', 'Nagole', np.nan, 'Pune']
           }
#dictionary to dataframe
d1 = pd.DataFrame(diction1)
d1.to_csv('data1.csv', index = False)
```

```python
diction2 = {'Names': ['Shamitha', 'Lasya', 'Ridhi', None],
           'Age': [None,22,18, 30],
           'City': ['Hyd', 'Nagole', None, 'Pune']}
d2 = pd.DataFrame(diction2)
d2.to_csv('data2.csv', index = False)
```
