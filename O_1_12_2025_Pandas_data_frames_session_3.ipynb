```python
import numpy as np
import pandas as pd
```

```python
filepath = '/content/Visadataset.csv'
visa_df = pd.read_csv(filepath)
visa_df
```

```python
visa_df.dtypes
```

 series : key value pairs without col names/ labels

```python
#extracting only keys
visa_df.dtypes.keys()
```

```python
#converting to list for better readability
columns = visa_df.dtypes.keys().tolist()
columns
```

same for values

```python
visa_df.dtypes.values()
```

```python
visa_df.dtypes.values
```

```python
dtype = visa_df.dtypes.values.tolist()
dtype
```

iterating over a list and storing them both values in a list

```python
cat = []
numerical = []
for i,j in zip(columns, dtype):
  if j == 'object':
    cat.append(j)
  else:
    numerical.append(j)
print(cat)
print(numerical)
```

```python
len(cat)
```

```python
len(numerical)
```

```python
#list comprehension
cat = [j for i,j in zip(columns, dtype) if j == 'object']
num = [j for i,j in zip(columns, dtype) if j != 'object']
len(cat), len(num)
```
