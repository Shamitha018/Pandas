DATA FRAMES

Step 1: Import packages

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

Step 2: Create Data Frame

```python
fruits = ['Kiwi', 'Apple', 'Strawberry']
pd.DataFrame(fruits)
```

Step 3: Change the default rows(index) and column(columns) names

```python
pd.DataFrame(fruits, ['Row1', 'Row2', 'Row3'])
```

Step 4: Add column Names

changing the order of keyword arguments, if keywords are not given then the parameters are considered as positional args

```python
pd.DataFrame(fruits, columns= ['Fruit_Names'],index=['Row1', 'Row2', 'Row3'])
```

```python
pd.DataFrame(fruits, [0,0,0],['Names'])
```

if only comma is given to skip second position, invalid syntax error is encountered

```python
pd.DataFrame(fruits, ,['names'])
```

Step 6: Multiple columns using zip function

```python
names = ['Ramesh', 'Suresh', 'Sathish']
age = [20,22,24]
pd.DataFrame(zip(names,age))
```
