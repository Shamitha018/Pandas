```python
uni = visa['continent'].unique()
uni
```

```python
l = []
for i in uni:
  con = visa['continent'] == i
  l.append(len(visa[con]))

```

```python
ft = pd.DataFrame(zip(uni,l), columns= ['class', 'frequency'])
ft.to_csv('Continent_frequency.csv', index= True)
```

values_counts

```python
visa['continent'].value_counts()
```

```python
labels = visa['continent'].value_counts
count = visa['continent'].value_counts
labels, count
```

```python
#bar chart
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
colors = ['pink', 'blue', 'orange', 'yellow', 'purple','green']
plt.bar(labels, count, color = colors)
plt.xlabel('Classes')
plt.ylabel('Frequency')
plt.title('Continent_barchart')
plt.savefig('continent_barchart.jpg')
plt.show() #after plt.show() no command will run
```
